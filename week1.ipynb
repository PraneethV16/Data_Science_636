```python
import numpy as np
import pandas as pd
df=pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
    })
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Department'].fillna(df['Department'].mode()[0],inplace=True)
print(df)
```

```python
#Forward fill
df=pd.DataFrame({'Age':[25,30,np.nan,40,35],'Deparment':['HR','Finance','Finance',np.nan,'IT']
              })
df_ffill=df.copy()
df_ffill.fillna(method='ffill',inplace=True)
print(df_ffill)
```

```python
#backward fill
df=pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})
df_bfill=df.copy()
df_bfill.fillna(method='bfill',inplace=True)
print(df_bfill)
```

```python
df=pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})
df_drop=df.dropna()
print("Drop rows:",df_drop)
```

```python
df=pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})
df_drop=df.dropna(axis=1)
print("Drop cols:",df_drop)
```

```python
df=pd.DataFrame({
    'ID':[1,2,2,3,4,4],
    'Name':['Alice','Bob','Bob','Charlie','David','David'],
    'Age':[25,30,30,35,40,40]
})
df_R=df.drop_duplicates()
print(df_R)

```

```python
df=pd.DataFrame({
    'Date':['2025-01-05','05/01/2025','Jan 5, 2025','2025.01.05']
})
df['Date']=pd.to_datetime(df['Date'],errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```
